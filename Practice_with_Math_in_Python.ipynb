## 1. Simple Math with `NumPy`
Introducing basic mathematical operations using `NumPy`. `NumPy` or Numerical Python, allows users to perfom math functions and apply them to arrays. 

```python
# we can always change our imports to represent anything we want, including 'np'
import numpy as np
```

```python
# Simple math operations
a = 10
b = 5

# take the sum
sum_ab = np.add(a, b)

# subtract two digits 
diff_ab = np.subtract(a, b)

# multiply
prod_ab = np.multiply(a, b)

# divide 
div_ab = np.divide(a, b)
```

```python
# lets show the results, in the cell above we have only *stored* them
# side note: there are three ways to print values

# 0. this one is doing the calculation in-line
print('Using method #0 ... ')
print('Sum', (np.add(a, b))) # and so on... 

# 1. calling the value directly
print('Using method #1 ... ')
print('Sum:',sum_ab)
print('Difference:', diff_ab)

# 2. using 'f-strings'
print('Using method #2 ... ')
print(f'Product: {prod_ab}')
print(f'Division: {div_ab}')
```

### Dabbling with pseudocode and boolean logic

```python
# i want a code that will calculate some value from two input values
# i want my calcultor to interpret my inputs and the input math calculation selected 
# how would i do that? 


# step 1: create a ... function
# step 2: create a ... condition
# step 3: create a(n) ... expression 
```

```python
# try this on your own ... 
```

```python
# Step 1: Create a function to perform the calculation
def calculate(input1, input2, operation):
    # Step 2: Create a condition to check the operation selected
    if operation == 'add':
        # Step 3: Create an expression for addition
        result = input1 + input2
    elif operation == 'subtract':
        # Step 3: Create an expression for subtraction
        result = input1 - input2
    elif operation == 'multiply':
        # Step 3: Create an expression for multiplication
        result = input1 * input2
    elif operation == 'divide':
        # Step 3: Create an expression for division, with a check for division by zero
        if input2 != 0:
            result = input1 / input2
        else:
            result = 'Error: Division by zero'
    else:
        result = 'Error: Invalid operation'
    
    return result
    #return print(f'{operation} {input1} from {input2}, and the result equals {result}')

```

```python
# Example usage:
input1 = 10
input2 = 10
operation = 'divide'
print(calculate(input1, input2, operation))  # Output: should be? 
```

Now we will put everything together

```python
from ipywidgets import interactive, FloatSlider, Text
import ipywidgets as widgets


# function to compute Z based on the given trigonometric functions
def calculate_Z(x, y, func_x, func_y):
    XX, YY = np.meshgrid(x, y)
    Z = eval(f'np.{func_x}(XX)') + eval(f'np.{func_y}(YY)')
    return XX, YY, Z

# interactive function to update the plot
def update_plot(x_min, x_max, y_min, y_max, func_x, func_y):
    x = np.linspace(x_min, x_max, 100)
    y = np.linspace(y_min, y_max, 100)
    XX, YY, Z = calculate_Z(x, y, func_x, func_y)
    
    plt.figure(figsize=(8, 6))
    plt.contourf(XX, YY, Z, levels=50, cmap='viridis')
    plt.colorbar()
    plt.title(f"Contour plot of Z = {func_x}(XX) + {func_y}(YY)")
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

# interactive stuff 
# sliders for adjusting the x and y axis limits
x_min_slider = FloatSlider(value=-2*np.pi, min=-10, max=0, step=0.1, description='x min')
x_max_slider = FloatSlider(value=2*np.pi, min=0, max=10, step=0.1, description='x max')
y_min_slider = FloatSlider(value=-2*np.pi, min=-10, max=0, step=0.1, description='y min')
y_max_slider = FloatSlider(value=2*np.pi, min=0, max=10, step=0.1, description='y max')

# text boxes for changing the trigonometric functions
func_x_text = Text(value='cos', description='function for x')
func_y_text = Text(value='sin', description='function for y')

# create interactive plot
interactive_plot = interactive(update_plot, x_min=x_min_slider, x_max=x_max_slider, 
                               y_min=y_min_slider, y_max=y_max_slider, 
                               func_x=func_x_text, func_y=func_y_text)

# display the interactive plot
widgets.VBox([interactive_plot])
```

```python
import matplotlib.pyplot as plt

plt.hist(sum_throws, bins=range(number_dice, number_dice*type_dice + 2), rwidth=0.8, align='left')
plt.xlabel('Sum of dice throws')
plt.ylabel('Number of throws')
plt.title(f'Histogram for throw of {number_dice} {type_dice}-sided dice')
plt.show()
```

```python
from ipywidgets import interact, IntSlider

def simulate_dice_throws(number_dice, number_throws):
    type_dice = 6
    dice_throws = np.random.randint(1, type_dice + 1, (number_throws, number_dice))
    sum_throws = np.sum(dice_throws, axis=1)

    plt.hist(sum_throws, bins=range(number_dice, number_dice * type_dice + 2), rwidth=0.8, align='left')
    plt.xlabel('Sum of dice throws')
    plt.ylabel('Number of throws')
    plt.title(f'Histogram for throw of {number_dice} {type_dice}-sided dice')
    plt.show()

interact(simulate_dice_throws,
         number_dice=IntSlider(min=1, max=20, step=1, value=3, description='Dice'),
         number_throws=IntSlider(min=1, max=10000, step=1, value=10000, description='Throws'))
```

## 4. Introduction to Pandas

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'Score': [88, 92, 95, 89]
}
df = pd.DataFrame(data)

print('DataFrame:')
print(df)
```

```python
# Basic operations on DataFrame
mean_age = df['Age'].mean()
max_score = df['Score'].max()

print('Mean Age:', mean_age)
print('Max Score:', max_score)
```

```python
# Adding a new column
passing_criteria = 90
df['Pass'] = df['Score'] > passing_criteria
print('Updated DataFrame:')
print(df)
```

```python
plt.grid()
plt.bar(df['Name'], df['Score'], color='slategray')
plt.axhline(passing_criteria, color='red')
plt.title('Graphing Test Scores \n Using Pandas')
plt.ylim(0,100)
```
